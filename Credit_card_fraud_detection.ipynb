```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset= pd.read_csv('creditcard.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset['Class'].value_counts()
```

```python
legit = dataset[dataset['Class'] == 0]
```

```python
fraud=dataset[dataset['Class']==1]
```

```python
print(legit)
```

```python
print(fraud)
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit_sample=legit.sample(n=134)
```

```python
balanced_dataset=pd.concat([legit_sample,fraud])
```

```python
balanced_dataset.shape
```

```python
balanced_dataset.head()
```

```python
balanced_dataset.tail()
```

```python
balanced_dataset['Class'].value_counts()
```

```python
X=balanced_dataset.drop(columns='Class',axis=1)
Y=balanced_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction=model.predict(X_train)
model_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(model_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
model_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(model_data_accuracy)
```
